## Part A — Plaintext bit flip, standard AES

**Hypothesis**: Under standard AES, a 1-bit flip in the plaintext should rapidly avalanche. After a few rounds, the **bit Hamming distance** should approach ~64/128 (about 50%) on average, and most bytes should differ.


```python
tables_a = Execute_Part_A()
df_a0 = pd.DataFrame([r.__dict__ for r in tables_a[0]])
df_a0
```

## Part B — Plaintext bit flip, ShiftRows disabled

**Hypothesis**: Disabling ShiftRows should reduce diffusion across columns. We expect smaller Hamming distances and slower growth across rounds versus Part A.


```python
tables_b = Execute_Part_B()
df_b0 = pd.DataFrame([r.__dict__ for r in tables_b[0]])
df_b0
```

## Part C — Plaintext bit flip, modified MixColumns

**Hypothesis**: Using the weaker custom MixColumns matrix should reduce diffusion strength, causing slower avalanche and more structure than Part A.


```python
tables_c = Execute_Part_C()
df_c0 = pd.DataFrame([r.__dict__ for r in tables_c[0]])
df_c0
```

## Part D — Key bit flip (fixed zero plaintext), repeat A/B/C configs

**Hypothesis**: A single key-bit flip should also avalanche via key expansion and AddRoundKey. Differences should appear immediately and then grow across rounds; diffusion quality should mirror Parts A/B/C depending on the config.


```python
tables_d = Execute_Part_D()
dfs_d = {k: pd.DataFrame([r.__dict__ for r in v]) for k, v in tables_d.items()}
list(dfs_d.keys())
```
